# CICDDoS2019 Dataset ‚Äî Feature Description

Below is a detailed explanation of each feature in the dataset.
The goal is to help you understand what each variable represents and how it relates to detecting DDoS attacks in network traffic.


##  **Basic Flow Information**

| Column            | Meaning                                                   |
| ----------------- | --------------------------------------------------------- |
| **Protocol**      | Network protocol used in the flow (e.g., TCP, UDP, ICMP). |
| **Flow Duration** | Total duration of the flow in microseconds.               |

---

##  **Packet Counts**

| Column                     | Meaning                                                         |
| -------------------------- | --------------------------------------------------------------- |
| **Total Fwd Packets**      | Number of packets sent from the source to the destination.      |
| **Total Backward Packets** | Number of packets sent from the destination back to the source. |

---

##  **Packet Size (Forward and Backward)**

| Column                       | Meaning                                       |
| ---------------------------- | --------------------------------------------- |
| **Fwd Packets Length Total** | Total size of all forward packets.            |
| **Bwd Packets Length Total** | Total size of all backward packets.           |
| **Fwd Packet Length Max**    | Maximum packet size in the forward direction. |
| **Fwd Packet Length Min**    | Minimum packet size in the forward direction. |
| **Fwd Packet Length Mean**   | Average forward packet size.                  |
| **Fwd Packet Length Std**    | Standard deviation of forward packet size.    |
| **Bwd Packet Length Max**    | Maximum backward packet size.                 |
| **Bwd Packet Length Min**    | Minimum backward packet size.                 |
| **Bwd Packet Length Mean**   | Average backward packet size.                 |
| **Bwd Packet Length Std**    | Standard deviation of backward packet size.   |

---

##  **Flow Rates**

| Column             | Meaning                                             |
| ------------------ | --------------------------------------------------- |
| **Flow Bytes/s**   | Byte rate of the whole flow (bytes per second).     |
| **Flow Packets/s** | Packet rate of the whole flow (packets per second). |
| **Fwd Packets/s**  | Forward transmission packet rate.                   |
| **Bwd Packets/s**  | Backward transmission packet rate.                  |

---

##  **Inter-Arrival Time (IAT) Measurements**

IAT = time between consecutive packets.

### Flow-level IATs:

| Column            | Meaning                            |
| ----------------- | ---------------------------------- |
| **Flow IAT Mean** | Average time between all packets.  |
| **Flow IAT Std**  | Variation in time between packets. |
| **Flow IAT Max**  | Longest gap between packets.       |
| **Flow IAT Min**  | Shortest gap between packets.      |

### Forward IATs:

| Column            | Meaning                                       |
| ----------------- | --------------------------------------------- |
| **Fwd IAT Total** | Total inter-arrival time for forward packets. |
| **Fwd IAT Mean**  | Average forward packet IAT.                   |
| **Fwd IAT Std**   | Variation in forward IAT.                     |
| **Fwd IAT Max**   | Maximum forward IAT.                          |
| **Fwd IAT Min**   | Minimum forward IAT.                          |

### Backward IATs:

| Column            | Meaning                         |
| ----------------- | ------------------------------- |
| **Bwd IAT Total** | Total IAT for backward packets. |
| **Bwd IAT Mean**  | Average backward IAT.           |
| **Bwd IAT Std**   | Variation in backward IAT.      |
| **Bwd IAT Max**   | Maximum backward IAT.           |
| **Bwd IAT Min**   | Minimum backward IAT.           |

---

##  **Flag Counters (TCP)**

| Column             | Meaning                                   |
| ------------------ | ----------------------------------------- |
| **Fwd PSH Flags**  | Number of forward packets with PSH flag.  |
| **Bwd PSH Flags**  | Number of backward packets with PSH flag. |
| **Fwd URG Flags**  | Number of forward packets with URG flag.  |
| **Bwd URG Flags**  | Number of backward packets with URG flag. |
| **FIN Flag Count** | Count of FIN flags in the flow.           |
| **SYN Flag Count** | Count of SYN flags in the flow.           |
| **RST Flag Count** | Count of RST flags in the flow.           |
| **PSH Flag Count** | Count of PSH flags in the flow.           |
| **ACK Flag Count** | Count of ACK flags in the flow.           |
| **URG Flag Count** | Count of URG flags in the flow.           |
| **CWE Flag Count** | Count of CWE flags.                       |
| **ECE Flag Count** | Count of ECE flags.                       |

Flags are very important for detecting abnormal behavior in DDoS attacks.

---

## üîß **Header Features**

| Column                | Meaning                              |
| --------------------- | ------------------------------------ |
| **Fwd Header Length** | Forward packet header total length.  |
| **Bwd Header Length** | Backward packet header total length. |

---

## üìê **Packet Length Statistics**

| Column                     | Meaning                              |
| -------------------------- | ------------------------------------ |
| **Packet Length Min**      | Smallest packet seen in flow.        |
| **Packet Length Max**      | Largest packet seen in flow.         |
| **Packet Length Mean**     | Average packet size.                 |
| **Packet Length Std**      | Packet size variability.             |
| **Packet Length Variance** | Statistical variance of packet size. |

---

##  **Flow Ratios and Averages**

| Column                   | Meaning                                     |
| ------------------------ | ------------------------------------------- |
| **Down/Up Ratio**        | Ratio of download to upload packets.        |
| **Avg Packet Size**      | Average size of all packets.                |
| **Avg Fwd Segment Size** | Average segment size in forward direction.  |
| **Avg Bwd Segment Size** | Average segment size in backward direction. |

---

##  **Bulk Features**

"Bulk" features represent groups of packets sent consecutively.

### Forward Bulk:

| Column                   | Meaning                    |
| ------------------------ | -------------------------- |
| **Fwd Avg Bytes/Bulk**   | Average bytes per bulk.    |
| **Fwd Avg Packets/Bulk** | Average packets per bulk.  |
| **Fwd Avg Bulk Rate**    | Rate of bulk transmission. |

### Backward Bulk:

| Column                   | Meaning                            |
| ------------------------ | ---------------------------------- |
| **Bwd Avg Bytes/Bulk**   | Average bytes per backward bulk.   |
| **Bwd Avg Packets/Bulk** | Average packets per backward bulk. |
| **Bwd Avg Bulk Rate**    | Backward bulk rate.                |

---

##  **Subflow Features**

| Column                  | Meaning                               |
| ----------------------- | ------------------------------------- |
| **Subflow Fwd Packets** | Number of packets in forward subflow. |
| **Subflow Fwd Bytes**   | Forward subflow bytes.                |
| **Subflow Bwd Packets** | Number of backward subflow packets.   |
| **Subflow Bwd Bytes**   | Backward subflow bytes.               |

---

##  **Window Features**

| Column                   | Meaning                           |
| ------------------------ | --------------------------------- |
| **Init Fwd Win Bytes**   | Initial forward TCP window size.  |
| **Init Bwd Win Bytes**   | Initial backward TCP window size. |
| **Fwd Act Data Packets** | Forward packets with actual data. |
| **Fwd Seg Size Min**     | Minimum forward segment size.     |

---

##  **Active/Idle Times**

Represent user or packet activity patterns.

| Column          | Meaning                           |
| --------------- | --------------------------------- |
| **Active Mean** | Average time the flow was active. |
| **Active Std**  | Activity variation.               |
| **Active Max**  | Maximum active time.              |
| **Active Min**  | Minimum active time.              |
| **Idle Mean**   | Average time the flow was idle.   |
| **Idle Std**    | Variation in idle time.           |
| **Idle Max**    | Maximum idle time.                |
| **Idle Min**    | Minimum idle time.                |

---

##  **Target Labels**

| Column    | Meaning                                       |
| --------- | --------------------------------------------- |
| **Label** | Text label of attack type (can be redundant). |
| **Class** | Binary target: **Attack** or **Benign**.      |

### Variance

```python
# Select numeric columns only
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Compute variance
variance = df[numeric_cols].var().sort_values()

for col in variance.index:
    if variance[col] < 1:
        print(f"Column '{col}' variance: {variance[col]:.6f}")

```

 **Variance-Based Feature Reduction**

After computing the variance of all numerical features, we applied a variance threshold to identify columns that do not contribute relevant information to the model.
Features with extremely low variance remain almost constant across the entire dataset, which means they do not help the classifier distinguish between benign and attack flows.

The following features showed very low variance and were therefore removed:

* `SYN Flag Count`
* `RST Flag Count`
* `CWE Flag Count`
* `URG Flag Count`
* `ACK Flag Count`

These columns represent sparse TCP flag indicators that rarely appear in the CICDDoS2019 dataset. Because their values are almost always zero, they do not provide useful discriminative power and can add noise or slow down training.


### ‚≠ê Exception: `Down/Up Ratio`

Although `Down/Up Ratio` also exhibited variance below 1, we **did not remove this feature**.
This characteristic captures the relationship between downstream and upstream packet flow and is **highly relevant for understanding DDoS attacks**, especially high-volume floods where traffic is predominantly unidirectional.

Keeping this feature helps the model identify:

* asymmetric traffic patterns
* unusual forward/backward flow behavior
* typical signatures of volumetric DDoS attacks


By removing low-variance features and retaining only the informative ones, we reduced noise and improved the stability of the deep learning model.
This step contributes to a more compact and meaningful feature space, supporting better performance for time-series architectures such as RNNs and Transformers.




```python
# ==========================================================
# Drop Low-Variance Columns
# Reason:
# These features have extremely low variance (<1),
# meaning they barely vary and have no predictive power.
# They also create noise and instability in deep models.
# ==========================================================

low_variance_drop = [
    'SYN Flag Count',
    'RST Flag Count',
    'CWE Flag Count',
    'URG Flag Count',
    'ACK Flag Count'
]

df = df.drop(columns=low_variance_drop, errors='ignore')

print("Dropped low-variance columns:")
print(low_variance_drop)
print("New shape:", df.shape)
```
